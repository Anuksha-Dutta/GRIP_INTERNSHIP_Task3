## Finding relations between the different variables of the dataset using different maps and visualisations

```python
# relationship analysis
correlation = data.corr()
```

```python
sns.heatmap(correlation, xticklabels = correlation.columns, yticklabels = correlation.columns
            , annot = True)
```

```python
sns.pairplot(correlation)
```

```python
sns.relplot(x= 'Sales', y='Profit', hue = 'Category', data= data)
```

```python
sns.relplot(x= 'Sales', y='Profit', hue = 'Ship Mode', data= data)
```

```python
sns.histplot(data['Profit'], bins=100)
```

```python
sns.catplot(x='Profit', kind='box', data=data)
```

```python
sns.catplot(x='Sales', kind='box', data=data)
```
